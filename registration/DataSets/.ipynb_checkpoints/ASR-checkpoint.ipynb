```python
anime_pivot = final_rating.pivot_table(columns='user_id', index='Name', values= 'rating')
```

```python
anime_pivot.fillna(0, inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
anime_sparse = csr_matrix(anime_pivot)
```

```python
anime_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(anime_sparse)
```

```python
distance, suggestion = model.kneighbors(anime_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=10)
```

```python
def recommend_anime(anime_name):
    anime_id = np.where(anime_pivot.index == anime_name)[0][0]
    distance, suggestion = model.kneighbors(anime_pivot.iloc[anime_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            anime = anime_pivot.index[suggestion[i]]
            for j in anime:
                if j == anime_name:
                    print(f"You searched '{anime_name}'\n")
                    print("The suggestion anime are: \n")
                else:
                    print(j)
```

```python
anime_name= 'Naruto'
recommend_anime(anime_name)
```

```python
import numpy as np

def recommend_anime_with_confidence(anime_name):
    anime_id = np.where(anime_pivot.index == anime_name)[0][0]
    distances, suggestions = model.kneighbors(anime_pivot.iloc[anime_id,:].values.reshape(1,-1), n_neighbors=6 )

    print(f"You searched '{anime_name}'\n")
    print("The suggested anime and their confidence levels are: \n")
    for distance, suggestion in zip(distances.flatten(), suggestions.flatten()):
        suggested_anime = anime_pivot.index[suggestion]
        confidence_level = 1 / (1 + distance)  # Calculating confidence level
        print(f"Anime: {suggested_anime}, Confidence Level: {confidence_level:.10f}")

anime_name= 'Naruto'
recommend_anime_with_confidence(anime_name)
```
